```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page=requests.get("https://www.imdb.com/list/ls055386972/")
soup = BeautifulSoup(page.text)

test_list = []

for i in soup.find_all('h3', class_ = 'lister-item-header'):
    test_list.append(i.text.strip())

movie_name = []
year = []
rating = []

for i in test_list:
    a = i.split()[1:-1]
    b = i.split()[-1]
    c = " ".join(a)
    movie_name.append(c)
    year.append(b)
    
    
for i in soup.find_all('span', class_="ipl-rating-star__rating"):
    rating.append(i.text)
    
dict_test = {"Movie Name" : movie_name, "Year" : year, "Rating" : rating[::23]}

df = pd.DataFrame(dict_test)
df.to_csv("50 film.csv")
df.head()
```
