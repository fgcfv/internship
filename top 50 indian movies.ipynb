```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page=requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
soup=BeautifulSoup(page.content)

text_file=[]
for i in soup.find_all('td',class_="titleColumn"):
    text_file.append(i.text.strip())
    
movie_name=[]
Release_year=[]
rating=[]

for i in text_file:
    a=i.split()[1:-1]
    b=i.split()[-1]
    c=" ".join(a)
    movie_name.append(c)
    Release_year.append(b)
    
for i in soup.find_all('td' ,class_="ratingColumn imdbRating"):
    rating.append(i.text.strip())
    
    
df=pd.DataFrame({"Movie Name" : movie_name, "Year" :Release_year,'Rating':rating})

df.head(51)
```
